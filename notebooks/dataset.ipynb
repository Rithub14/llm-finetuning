## cleaning the final dataset

```python
final_data.size
```

```python
final_data.shape
```

```python
final_data.columns
```

```python
final_data = final_data.dropna(subset=['instruction', 'response'])
```

```python
final_data = final_data[
    (final_data["instruction"].str.strip() != "") &
    (final_data["response"].str.strip() != "")
]
```

```python
final_data = final_data[
    (final_data["instruction"].str.len() > 5) &
    (final_data["response"].str.len() > 5)
]
```

```python
final_data.shape
```

```python
final_data = final_data.drop_duplicates(subset=["instruction", "response"])
```

```python
final_data.shape
```

```python
final_data = final_data.reset_index(drop=True)
```

```python
final_data.shape
```

```python
final_data_path = "../dataset/data.csv"
final_data.to_csv(final_data_path, index=False)
```
